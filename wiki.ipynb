## Data Frame 1:  
A partir de https://restcountries.com/, se hallaron los países sudamericanos y sus capitales, idiomas, longitud y latitud.

```python
import requests
import pandas as pd

url = "https://restcountries.com/v3.1/region/South%20America"

response = requests.get(url)
data = response.json()

if response.status_code == 200:
    countries = data

    table_data = []

    # Obtener información de cada país
    for country in countries:
        country_name = country["name"]["common"]
        capital = country.get("capital", "N/A")
        languages = ", ".join(country.get("languages", []))
        latlng = country.get("latlng", [])

        # Obtener latitud y longitud
        latitude = latlng[0] if len(latlng) >= 1 else "N/A"
        longitude = latlng[1] if len(latlng) >= 2 else "N/A"

        # Agregar los datos del país a la lista
        table_data.append({
            "País": country_name,
            "Capital": capital,
            "Idiomas": languages,
            "Latitud": latitude,
            "Longitud": longitude
        })

    # Convertir la lista en un DataFrame
    df1 = pd.DataFrame(table_data)

    # Imprimir el DataFrame
    print(df1)
else:
    print("Error al obtener los datos. Código de respuesta:", response.status_code)
```

## Data Frame 3:  
A partir de https://openweathermap.org/ se obtiene la temperatura de los países sudamericanos en kelvin.

```python
import requests
import pandas as pd

api_key = 'cf62addae9c1d0360bf97aa196222a0b'

url_countries = "https://restcountries.com/v3.1/region/South America"
response_countries = requests.get(url_countries)
data_countries = response_countries.json()

if response_countries.status_code == 200:
    countries = data_countries

    temperatures = []
    country_names = []

    for country in countries:
        country_name = country["name"]["common"]
        country_names.append(country_name)

        url_weather = f'http://api.openweathermap.org/data/2.5/weather?q={country_name}&appid={api_key}'
        response_weather = requests.get(url_weather)
        data_weather = response_weather.json()

        if 'main' in data_weather and 'temp' in data_weather['main']:
            temperature = data_weather['main']['temp']
            temperatures.append(temperature)
        else:
            temperatures.append(None)

    df3 = pd.DataFrame({'País': country_names, 'TemperaturaK': temperatures})
    print(df3)
else:
    print("Error al obtener los datos de los países. Código de respuesta:", response_countries.status_code)
```

### Unión de los tres data frames  

```python
import pandas as pd

# Supongamos que tienes tres DataFrames: df1, df2 y df3

# Realizar la unión de los DataFrames basados en la columna 'país'
merged_df = pd.merge(df1, df2, on='País', how='inner')
merged_df = pd.merge(merged_df, df3, on='País', how='inner')

# Imprimir el DataFrame resultante
print(merged_df)
```
