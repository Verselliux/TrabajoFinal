```python
import requests

api_key = 'cf62addae9c1d0360bf97aa196222a0b'
countries = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic', 'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela']
month = 7  # Mes deseado (ejemplo: julio)

for country in countries:
    url = f'http://api.openweathermap.org/data/2.5/forecast?q={country}&appid={api_key}&units=metric'
    response = requests.get(url)
    data = response.json()
    
    city_name = data['city']['name']
    
    print(f'Temperaturas en {city_name} durante el mes {month}:')
    
    for item in data['list']:
        date = item['dt_txt']
        temperature = item['main']['temp']
        
        if int(date.split('-')[1]) == month:
            print(f'Fecha: {date}, Temperatura: {temperature} °C')
```

```python
import requests

api_key = 'cf62addae9c1d0360bf97aa196222a0b'
countries = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic', 'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela']

for country in countries:
    url = f'http://api.openweathermap.org/data/2.5/weather?q={country}&appid={api_key}'
    response = requests.get(url)
    data = response.json()

    temperature = round(data['main']['temp'] - 273.15, 2)
    country_name = data['name']
    max_temp = round(data['main']['temp_max'] - 273.15, 2)
    min_temp = round(data['main']['temp_min'] - 273.15, 2)

    print(f'Temperatura actual en {country_name}: {temperature} °C')
    print(f'Temperatura máxima en {country_name}: {max_temp} °C')
    print(f'Temperatura mínima en {country_name}: {min_temp} °C')
    print()
```
